```python
# removinge outliers
for col in ['fare_amount','trip_distance','duration']:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1

    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR

    df=df[(df[col]>=lower_bound) & (df[col]<=upper_bound)]
```

```python
# using t -test
card_sample1=df[df['payment_type']== 'Card']['fare_amount'].sample(100)
cash_sample1=df[df['payment_type']== 'Cash']['fare_amount'].sample(100)
```

```python
t_stats,p_value=stats.ttest_ind(a=card_sample1,b=cash_sample1,equal_var=False)
if p_value<0.05:
  print("reject null hypothesis")
else:
  print("accept null hypothesis")
print("t_stats",t_stats)
print("p_value",p_value )
```
